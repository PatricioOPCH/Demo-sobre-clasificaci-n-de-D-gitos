```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()


x_train, x_test = x_train / 255.0, x_test / 255.0

y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=20, batch_size=30)

test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Precisión en el conjunto de prueba: {test_accuracy}')
```

Importación del Dataset MNIST

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

Mostramos imagenes del dataset para su observación

```python
import matplotlib.pyplot as plt
num_imagenes_a_mostrar = 8

for i in range(num_imagenes_a_mostrar):
    plt.subplot(1, num_imagenes_a_mostrar, i + 1)
    plt.imshow(x_train[i], cmap='gray')  # Mostrar la imagen en escala de grises
    plt.title({y_train[i]})  # Mostrar la etiqueta correspondiente
    plt.axis('off')  # Ocultar los ejes
plt.show()
```

Preprocesamiento de datos

```python
# Reajusta las imágenes de entrenamiento para tener la dimensión del canal (1) necesario para la convolución
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)

# Reajusta las imágenes de prueba para tener la dimensión del canal (1) necesario para la convolución
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

# Normaliza los valores de píxeles de las imágenes dividiéndolos por 255 para que estén en el rango de 0 a 1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Convierte las etiquetas de entrenamiento a su representación one-hot encoding para el entrenamiento del modelo
y_train = tf.keras.utils.to_categorical(y_train, 10)
# Convierte las etiquetas de prueba a su representación one-hot encoding para el entrenamiento del modelo
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
# Crea un generador de aumento de datos utilizando ImageDataGenerator de Keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=10,         # Define el rango de rotación para las imágenes (en grados)
    width_shift_range=0.1,     # Define el rango de desplazamiento horizontal de las imágenes
    height_shift_range=0.1,    # Define el rango de desplazamiento vertical de las imágenes
    zoom_range=0.1             # Define el rango de zoom de las imágenes
)
```

Selección y configuración del Modelo 

Este modelo es una red neuronal convolucional (CNN) que utiliza capas convolucionales, capas de pooling (agrupación), capas densas y funciones de activación específicas:

* Capa Conv2D: Esta capa realiza convolución con 32 filtros de 3x3, utilizando la función de activación ReLU (Rectified Linear Unit). La capa toma imágenes de entrada de tamaño 28x28 con un solo canal (escala de grises).

* Capa MaxPooling2D: Reduce la dimensionalidad y extrae características mediante el agrupamiento máximo con una ventana de 2x2.

* Capa Flatten: Convierte la salida de las capas anteriores en un vector unidimensional para alimentar capas densas.

* Capas Dense: Son capas completamente conectadas. La primera tiene 128 neuronas con activación ReLU y la segunda tiene 10 neuronas con activación Softmax, que es ideal para problemas de clasificación, ya que devuelve probabilidades para cada clase.

Este tipo de modelo es efectivo para tareas de clasificación de imágenes, como en el conjunto de datos MNIST (dígitos escritos a mano), ya que las capas convolucionales aprenden patrones visuales, mientras que las capas densas realizan la clasificación. El uso de ReLU y Softmax como funciones de activación es común debido a su eficacia en tareas de aprendizaje automático.

```python
# Creación de un modelo secuencial utilizando Keras
model = tf.keras.Sequential([
    # Capa de convolución con 32 filtros de 3x3, activación ReLU y forma de entrada (28x28x1)
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    
    # Capa de Max Pooling con ventana de 2x2 para reducir dimensionalidad y extraer características
    tf.keras.layers.MaxPooling2D((2, 2)),
    
    # Capa de aplanamiento para convertir la salida en un vector unidimensional
    tf.keras.layers.Flatten(),
    
    # Capa densa con 128 neuronas y activación ReLU para aprendizaje de características
    tf.keras.layers.Dense(128, activation='relu'),
    
    # Capa densa final con 10 neuronas (clases) y activación Softmax para clasificación
    tf.keras.layers.Dense(10, activation='softmax')
])
```

Compilación de modelo

Donde definieremos : 
        model.compile: Este paso compila el modelo para la fase de entrenamiento, estableciendo las configuraciones esenciales para la optimización y evaluación del modelo.

* optimizer='adam': Se elige el optimizador 'Adam'. Adam es un optimizador de descenso de gradiente estocástico muy popular en el aprendizaje profundo. Ajusta automáticamente su tasa de aprendizaje durante el entrenamiento.

* loss='categorical_crossentropy': En problemas de clasificación con varias clases (como el caso de 10 clases en el MNIST), la función de pérdida 'categorical_crossentropy' se usa comúnmente. Esta función calcula la diferencia entre las distribuciones de probabilidad predichas por el modelo y las distribuciones reales de las etiquetas.

* metrics=['accuracy']: Esta configuración define la métrica que se utilizará para evaluar el modelo durante el entrenamiento. En este caso, se usa 'accuracy' (precisión), que mide la proporción de predicciones correctas en relación con el total.

Este paso de compilación es crucial antes de entrenar el modelo, ya que establece cómo se calcularán las actualizaciones de los pesos durante el entrenamiento (optimizador), cómo se medirá la diferencia entre las predicciones y las etiquetas reales (función de pérdida) y qué métricas se utilizarán para evaluar el rendimiento del modelo.

```python
# Compila el modelo con la configuración definida
model.compile(optimizer='adam',                   # Optimizador Adam para el entrenamiento
              loss='categorical_crossentropy',    # Función de pérdida para problemas de clasificación múltiple
              metrics=['accuracy'])              # Métrica para evaluar la precisión del modelo

```

Entrenamiento del Modelo

Donde: 
    
* datagen.fit(x_train): Prepara el generador de aumento de datos (datagen) según el conjunto de entrenamiento (x_train). Esto establece los parámetros del generador para la generación de datos aumentados.

* model.fit(datagen.flow(x_train, y_train, batch_size=30), epochs=20): Entrena el modelo usando el generador de datos aumentados (datagen.flow). El método flow toma los datos de entrada (x_train) y las etiquetas correspondientes (y_train) con un tamaño de lote de 30. Este entrenamiento se lleva a cabo durante 20 épocas.

* test_loss, test_accuracy = model.evaluate(x_test, y_test): Evalúa el modelo con el conjunto de datos de prueba (x_test y y_test) y registra tanto la pérdida como la precisión del modelo.
* print(f'Precisión en el conjunto de prueba: {test_accuracy}'): Muestra la precisión del modelo en el conjunto de prueba, lo que indica cuántas predicciones son correctas en relación con el total de predicciones realizadas en el conjunto de datos de prueba.

```python
datagen.fit(x_train)
# Ajusta el generador de aumento de datos según el conjunto de entrenamiento

model.fit(datagen.flow(x_train, y_train, batch_size=30), epochs=20)
# Entrena el modelo utilizando el generador de datos aumentados, 
# flujo de datos, y un tamaño de lote de 30, durante 20 épocas.

test_loss, test_accuracy = model.evaluate(x_test, y_test)
# Evalúa el modelo en el conjunto de datos de prueba y registra la pérdida y precisión.

print(f'Precisión en el conjunto de prueba: {test_accuracy}')
# Imprime la precisión del modelo en el conjunto de prueba.
```
